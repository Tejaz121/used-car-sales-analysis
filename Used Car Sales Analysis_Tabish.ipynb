```python
data[car_prices_in_rupee]=data[car_prices_in_rupee].str.replace("lakh","",reged=false).astype(float)*100000
```

```python
data['car_prices_in_rupee'] = data['car_prices_in_rupee'].str.replace(" Lakh", "", regex=False).astype(float) * 100000
```

```python
data['car_prices_in_rupee'] = np.where(
    data['car_prices_in_rupee'].str.contains('Crore'),
    data['car_prices_in_rupee'].str.replace(' Crore', '', regex=False).astype(float) * 1_00_00_000,
    data['car_prices_in_rupee'].str.replace(' Lakh', '', regex=False).astype(float) * 1_00_000
)
```

```python
def convert_price(price):
    if 'Crore' in price:
        return float(price.replace(' Crore', '')) * 10000000
    elif 'Lakh' in price:
        return float(price.replace(' Lakh', '')) * 100000
    else:
        return None  # or 0 or np.nan, depending on your use case

data['car_prices_in_rupee'] = data['car_prices_in_rupee'].apply(convert_price)
```

```python
print(data[['car_name', 'car_prices_in_rupee']].head())
```

```python
# We defined a function that checks if the price contains "Crore" or "Lakh", removes the text, converts the numeric part to float, and multiplies it by the corresponding value (10 million for Crore and 100 thousand for Lakh).
```

```python
print(data.dtypes)
```

```python
# Convert kms_driven to numeric (remove 'kms' and commas)
data['kms_driven'] = data['kms_driven'].str.replace(' kms', '', regex=False).str.replace(',', '', regex=False).astype(float)

# Convert Seats to numeric (remove 'Seats' and convert to numeric)
data['Seats'] = data['Seats'].str.replace(' Seats', '', regex=False).astype(int)
```

```python
import datetime
current_year = datetime.datetime.now().year
data['Car_Age'] = current_year - data['manufacture']
```
